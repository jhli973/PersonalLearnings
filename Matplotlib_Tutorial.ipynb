```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import random
```

```python
# simple plot
plt.plot([1,2,3], [1,4,9])
plt.show()
```

### 1. Legends, Title, Subtitle, Labels

```python
x = np.array([1,2,3,4])
y1 = np.power(x,2)
y2 = np.exp(x)
plt.plot(x, y1, label="Power")
plt.plot(x, y2, label="Exponential")
plt.title('Simple plot\nSubtitle')
plt.xlabel('X values')
plt.ylabel('Calculated y value')
plt.legend()
plt.show()
```

### 2. Histogram, bar chart, scatter and stack plots

```python
# set seed
random.seed(10)

x1 = np.array([1,2,3,4])
# random integer generator
y1 = np.random.random_integers(1, 10, 4)
x2 = np.array([6,7,8,9])
y2 = np.random.random_integers(1, 10, 4)
plt.bar(x1,y1, label = "Customer1", color = 'b')
plt.bar(x2,y2, label = "Customer2", color='c')
plt.title("Bar chart")
plt.xlabel('Customers')
plt.ylabel('Values')
plt.show()
```

```python
# set seed
random.seed(10)

# random integer generator
mu, sigma = 0, 1.0

# generate normal distributed 100 number with 0 average and 1 sdt
x = np.random.normal(mu, sigma, 1000)

plt.hist(x, bins=12, normed=True, linewidth=2)
plt.title("Histogram")
plt.ylabel('Frequency')
plt.show()
```

```python
# set seed
random.seed(10)
x = np.random.random_integers(1, 10, 10)
y = np.random.random_integers(1, 10, 10)

plt.scatter(x,y, color='r', marker='x')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('random scatter')
plt.show()
```

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,10,7]
eating   = [2,3,4,2,3]
working  = [7,8,7,4,3]
playing  = [8,5,7,8,11]

plt.plot([], [], color='r', label='Sleeping', linewidth=5)
plt.plot([], [], color='g', label='Eating', linewidth=5)
plt.plot([], [], color='m', label='Working', linewidth=5)
plt.plot([], [], color='b', label='Playing', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors= ['r', 'g', 'm', 'b'])

plt.xlabel('weekdays')
plt.ylabel('hours')
plt.title('Summary of time spending')

# put legen outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# pie chart

slices = [7,2,6,9]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c', 'm','g','b']

plt.pie(slices, 
        labels=activities, 
        colors= cols, 
        shadow=True, 
        explode=(0, 0.2,0,0), # extend out the 2nd value
        autopct= '%1.1f%%'  # add percentage
       )
plt.title("Summary of time spending")
plt.show()
```

```python
from pandas.io.data import DataReader
import datetime
import time

end = datetime.date.today()
start = end - datetime.timedelta(days=3650)

google_stock = DataReader('GOOG', 'yahoo', start, end)
amazon_stock = DataReader('AMZN', 'yahoo', start, end)
google_stock.head()
```

```python
plt.figure(figsize=(10,7))

plt.plot(google_stock.index.to_pydatetime(), google_stock['Open'], label="Google", color='m')
plt.plot(amazon_stock.index.to_pydatetime(), amazon_stock['Open'], label="Amazon", color='b')

plt.title("Stock charts")
plt.ylabel("Open Price")
plt.xlabel("Years")
plt.legend()
plt.show()
```

### Use style

```python
from matplotlib import style

print (plt.style.available)
```

```python
# test different styles

#style.use('dark_background')
#style.use('seaborn-poster')
#style.use('ggplot')
style.use('seaborn-white')
#style.use('fivethirtyeight')
#style.use('seaborn-darkgrid')

plt.figure(figsize=(10,7))

plt.plot(google_stock.index.to_pydatetime(), google_stock['Open'], label="Google", color='m')
plt.plot(amazon_stock.index.to_pydatetime(), amazon_stock['Open'], label="Amazon", color='b')

ax = plt.gca()
ax.grid(True)  # add grid

ax.set_ylim([0, 1600])
ax.set_yticks([0,400,800,1200, 1600])

# fill under the curve
ax.fill_between(google_stock.index, google_stock['Open'], facecolor='m', alpha=0.5)

# set label color as bllue
ax.yaxis.label.set_color('b')
ax.xaxis.label.set_color('b')

# set spine color
ax.spines['left'].set_color('k')
ax.spines['bottom'].set_color('k')

ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# add a holizontal line
ax.axhline(400, color='g', linewidth=3)

# set tick color -- you can refer color-hex.com
ax.tick_params(axis='x', colors='#042C62') 


plt.title("Stock charts")
plt.ylabel("Open Price")
plt.xlabel("Years")
plt.legend()
plt.show()
```
