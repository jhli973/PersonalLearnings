```python
def fibonacci(n):
    
    fib_lst = []
    x = 0
    fib = 0
    
    while len(fib_lst) < n:
        
        if x == 0:
            fib = 0
        elif x == 1:
            fib = 1
        else:
            fib += x
        x += 1
        fib_lst.append(fib)
    return fib_lst
           
```

```python
cube = lambda x: x ** 3

def fib(x):
    if x==1:
        return 0
    if x == 2:
        return 1
    return fib(x-1) + fib(x-2)

def fibonacci(n):
    # return a list of fibonacci numbers
    fib_lst = []

    if n == 1:
        fib_lst.append(0)
    else:
        for i in range(1, n+1):
            fib_lst.append(fib(i))
    return fib_lst
```

```python
fibonacci(5)
```
