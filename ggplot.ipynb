## About ggplot from yhat, please refer [here](http://ggplot.yhathq.com/)

```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series

import ggplot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from ggplot import *
meat.head()
```

```python
g = ggplot(aes(x='date', y='beef'), data=meat)
g
```

```python
# add points
g + geom_point()
```

```python
# add a line 
g + geom_point() +geom_line()
```

```python
# add a trendline
g + geom_point() + geom_line() + stat_smooth(color='blue')
```

```python
# histogram
ggplot(aes(x='beef'), data=meat) + geom_histogram(binwidth=50)
```

```python
df = pd.DataFrame({
    "x": np.random.normal(0, 10, 1000),
    "y": np.random.normal(0, 10, 1000),
    "z": np.random.normal(0, 10, 1000)
})

df = pd.melt(df)  # stack into one column
df.head(5)
```

```python
ggplot(aes(x='value', fill='variable', color='variable'), data=df) + geom_histogram(binwidth=3, alpha=0.6)
```
