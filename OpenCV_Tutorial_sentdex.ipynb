## Image read and show

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
img = cv2.imread('elephant.jpg', cv2.IMREAD_UNCHANGED)  #IMREAD_GRAYSCALE or 0 #IMREAD_COLOR or 1; IMREAD_UNCHANGED or -1

#draw a line, on img, start point and end point and color and width
cv2.line(img, (0,400),(1000, 400), (255,255,255), 10)
#draw a rectangle
cv2.rectangle(img, (240, 20), (640, 400), (0, 0, 255), 5)
# draw a circle
cv2.circle(img, (700, 100), 50, (255, 0, 0), -1)
# draw  polylines
pts = np.array([[50,0], [20,10], [30,100], [70,30]], np.int32)
cv2.polylines(img, [pts], True, (0,255,255), 2)

# write text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, 'Giant King', (270,450), font, 2, (0,255,50), 2,cv2.LINE_AA)

cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.show()
````

````python
cv2.imwrite('elegray.png', img)
````

## Basic image operation

````python
img = cv2.imread('elephant.jpg', cv2.IMREAD_COLOR) 
print(img.shape)
# image pixel
px = img[300,200]
print(px)
img[50,50] = [255,255,255]
print(px)
````

````python
# region of image (ROI)

roi = img[0:100, 0:100]
print(roi)
cv2.imshow('roi', roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
img[0:100, 0:100] = [255,255,255]

# you will see a white square in the upper left corner
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# copy region
ele_head = img[40:400, 260:640]
img[140:500,620:1000] = ele_head
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
