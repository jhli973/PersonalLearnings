## Image read and show

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('elephant.jpg', cv2.IMREAD_GRAYSCALE)  #IMREAD_GRAYSCALE or 0 #IMREAD_COLOR or 1; IMREAD_UNCHANGED or -1
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.show()
```

```python
cv2.imwrite('elegray.png', img)
```
